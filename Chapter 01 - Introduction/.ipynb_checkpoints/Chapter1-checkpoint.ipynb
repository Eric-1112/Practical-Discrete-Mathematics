### The `sort()` function

```python
numbers = [3, 1, 4, 12, 8, 5, 2, 9]
names = ['Wyatt', 'Brandon', 'Kumar', 'Eugene', 'Elise']

# Apply the sort() function to the lists
numbers.sort()
names.sort()

# Display the output
print(numbers)
print(names)
```

### The `shuffle()` function

```python
import random

# Set a random seed so the code is reproducible
random.seed(1)

# Run the random.shuffle() function 5 times and display the outputs
for i in range(0,5):
    numbers = [3, 1, 4, 12, 8, 5, 2, 9]
    random.shuffle(numbers)
    print(numbers)
```

```python
import pandas as pd

P = [True,True,False,False]
Q = [True,False,True,False]

P_and_Q = [P and Q for P,Q in zip(P,Q)]
P_or_Q = [p or q for p,q in zip(P,Q)]
not_P = [not p for p in P]

truth_table = {
    'P' : P,
    'Q' : Q,
    'P AND Q' : P_and_Q,
    'P OR Q' : P_or_Q,
    'NOT P' : not_P,
}

df = pd.DataFrame(truth_table)
#Display the truth table
print(df)
```

Question;Request user input of a simple proposition statement.Create a truth table

```python
import pandas as pd

# Define the proposition
P = [True, False]

# Create the truth table
truth_table = pd.DataFrame({
    'P (The light is on)': P,
    '¬P (The light is off)': [not p for p in P]
})

# Display the truth table
truth_table
```

```python
import pandas as pd

P = [True,True,True,True,False,False,False,False]
Q = [True,True,False,False,True,True,False,False]
R = [True,False,True,False,True,False,True,False]

P_and_Q = [P and Q for P,Q in zip(P,Q)]
P_and_R = [P and R for P,R in zip(P,R)]
Q_and_R = [Q and R for Q,R in zip(Q,R)]
P_and_Q_and_R = [P and Q and R for P,Q,R in zip(P,Q,R)]
P_or_Q = [p or q for p,q in zip(P,Q)]
P_or_R = [p or r for p,r in zip(P,R)]
Q_or_R = [q or r for q,r in zip(Q,R)]
P_or_Q_or_R = [p or q or r for p,q,r in zip(P,Q,R)]
not_P = [not p for p in P]
not_Q = [not q for q in Q]
not_R = [not r for r in R]

truth_table = {
    'P' : P,
    'Q' : Q,
    'R' : R,
    'P AND Q' : P_and_Q,
    'P AND R' : P_and_R,
    'Q AND R' : Q_and_R,
    'P AND Q AND R' : P_and_Q_and_R,
    'P OR R' : P_or_R,
    'P OR Q' : P_or_Q,
    'Q OR R' : Q_or_R,
    'P OR Q OR R' : P_or_Q_or_R,
    'NOT P' : not_P,
    'NOT Q' : not_Q,
    'NOT R' : not_R,
}

df = pd.DataFrame(truth_table)
print(df)
```
